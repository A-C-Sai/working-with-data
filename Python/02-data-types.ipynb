# Mutability and Immutability

### Mutability

An object is said to be MUTABLE if and only if It allows us to perform changes / Updations / Modifications at ```Same Address```

### Immutability

An object is said to be IMMUTABLE if and only if It satisfies the following Properties:
- It never allows to modify at  Same Address ( It allows us to  do the changes and changed value placed into Different Address
- It never allows us to Perform "Item Assigment" (OR) does not support item assignment)

# Shallow Vs Deep Copy

copy module:
- copy.copy() One Level Deep, only references of nested child objects
- copy.deepcopy() full independent copy

### Shallow Copy

1) Intial Content of Both the Objects are SAME
2) The Memory Address (reference) of Both the Objects are DIFFERENT
3) Modifications are INDEPENDENT ( Whatever the changes we do on one object, It is not reflecting on anothetr object)

### Deep Copy

1) Intial Content of Both the Objects are SAME
2) The Memory Address (reference) of Both the Objects are SAME
3) Modifications are DEPENDENT ( Whatever the changes we do on one object, and  reflecting on anothetr object)
