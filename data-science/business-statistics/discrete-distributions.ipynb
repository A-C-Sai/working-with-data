# Binomial

A manufacturing company produces 10,000 plastic mugs per week. This companysupplies mugs to another company, which packages the mugs as part of picnic sets.<br>The second company randomly **samples 10 mugs** sent from the supplier. **If two orfewer of the sampled mugs are defective, the second company accepts the lot**.<br>
What is the probability that the lot will be accepted if the mug manufacturing companyactually is producing mugs that are:<br>
**10%** defective?<br>
**20%** defective?<br> 
**30%** defective?<br>**40%** defective?<br>

```python
n = 10
p1 = 0.1
p2 = 0.2
p3 = 0.3
p4 = 0.4
x = 2
```

```python
results = [[stats.binom.cdf(x,n,p1),stats.binom.mean(n,p1)],
           [stats.binom.cdf(x,n,p2),stats.binom.mean(n,p2)],
           [stats.binom.cdf(x,n,p3),stats.binom.mean(n,p3)],
           [stats.binom.cdf(x,n,p4),stats.binom.mean(n,p4)]]
```

```python
res_df = pd.DataFrame(results,columns=['Acceptance Probability','Expected Number of Defects'])
```

```python
res_df = pd.concat([res_df,pd.Series([p1,p2,p3,p4])],axis=1).rename(columns={0:'p'})
```

```python
res_df
```

```python
res_df.plot.bar(x='p',y='Acceptance Probability')
plt.show()
```

# Poisson

```python
lambda_ = 3.2
x = 7
```

```python
round((1 - stats.poisson.cdf(x,lambda_))*100,3)
```

If the bank has been averaging 3.2 customers every 4 minutes on weekday
afternoons, it is unlikely that more than 7 people would randomly arrive in any one
4-minute period. This answer indicates that more than 7 people would randomly
arrive in a 4-minute period only 1.69% of the time. Bank officers could use these
results to help them make staffing decisions.

**Always adjust the lambda value**<br>
if we expect 3.2 customers for every 4 minutes, we should expect twice as many, 6.4 customers every 8 minutes

```python
lambda_new = 6.4
x = 10
```

```python
round(stats.poisson.pmf(x,lambda_new)*100,3)
```

With large sample sizes and small values of p, we can approximate a binomial problem with a poisson distribution.<br>
While using the binomial will give the exact result, poission works as a good approximation.

```python
binom = [stats.binom.pmf(i,100000,0.000014) for i in range(20)]
poisson = [stats.poisson.pmf(i,100000*0.000014) for i in range(20)]
```

```python
binom_vs_poisson = pd.DataFrame({'binomial':binom,'poisson':poisson})
```

```python
binom_vs_poisson.plot.bar(rot=0)
```
