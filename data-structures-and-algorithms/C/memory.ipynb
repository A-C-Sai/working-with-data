```python
%%file memory.c
#include <stdio.h>
int main() {

    int n;
    printf("How many characters would you like to store: ");
    scanf("%d",&n);
    char A[n];

    printf("What are the characters you would like to store:\n");
    for(int i=0;i<n-1;i++) {
        printf("Enter a character: ");
        scanf(" %c",A+i);
    }
    A[n-1] = '\0';


    printf("%s",A);

    
    return 0;
}
```
